# Exercise 2. Watermarking
## Part A

### Import Liberaries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Watermarking Function that Calculate DCT of Two Images and Sum up them together

```python
def watermark(src, target, c):
    src = cv2.dct(src)
    target = cv2.dct(target)
    img = np.copy(src)
    img[0:target.shape[0], 0:target.shape[1]] += target * c
    return cv2.idct(img)
```

### Reading Original Image and Showing It

```python
barbara = cv2.imread("barbara.tif", cv2.IMREAD_GRAYSCALE).astype("float64")
plt.imshow(barbara, cmap='gray')
plt.title("Original Image")
plt.show()
```

### Creating the Target Mask

```python
mask = np.ones((300, 200))
mask[50:250, 50:150] = np.zeros((200, 100))
```

### Watermarking the mask into the barbara image with different coefficients

```python
c = [1, 10, 100, 1000]
for i in c:
    watermarked = watermark(barbara, mask, i)
    plt.imshow(watermarked, cmap='gray')
    plt.title(f"c = {i}")
    plt.show()
```
