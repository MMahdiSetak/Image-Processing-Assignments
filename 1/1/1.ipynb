# تمرین ۱: اصلاح هیستوگرام
تصویر همواپیما را در ورودی خوانده و نمایش دهید. هیستوگرام تصویر را محاسبه نمایید و نمایش دهید. با استفاده از روش
همسان سازی هیستوگرام، تصویر را نگاشت داده و آن را با تصویر اولیه مقایسه نمایید. مقایسه شامل تصاویر و هیستوگرامهای تصاویر
در قالب گزارش ارایه شود

### Import libraries and read and show the original image

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("Airplane.tiff",cv2.IMREAD_GRAYSCALE)
plt.imshow(image,cmap='gray')
```

### Calculating histogram and ploting it

```python
hist = cv2.calcHist([image],[0],None,[256],[0,255])
plt.plot(hist)
plt.show()

plt.hist(image.ravel(),bins=256,range=[0,255])
plt.show()
```

مشاهده میشود اکثر مقادیر پیکسل های دارای اعداد کوچک هستند و
اکثر مقادیر متمرکز در ناحیه صفر تا پنجاه هستند که باعث میشود تصویری تیره داشته باشیم
انتظار می رود با همسان سازی هیتسوگرام این مقدادیر در بازه صفر تا ۲۵۵ پخش شوند تا تصویر تا حدودی واضح تر شود

### Equlizing image histogram and showing the result

```python
equalized_image = cv2.equalizeHist(image)
plt.imshow(equalized_image,cmap='gray')
```

مشاهده میشود به علت همسان سازی گستره روشنایی تصویر افزایش یافته و نسبت به تصویر اولیه جزئیات بیشتر قابل مشاهده می باشد
در کل نیز می توان گفت کیفیت تصویر افزایش یافته

### Calculating equalized image and ploting the result

```python
hist = cv2.calcHist([equalized_image],[0],None,[256],[0,255])
plt.plot(hist)
plt.show()
```

با اعمال همسان سازی مقادیر پیکسل ها در بازه ۰ تا ۲۵۵ پخش می شوند که در نمودار قابل ملاحظه می باشد

### نتیجه گیری
با همسان سازی می توان در موارد مشابه کیفیت تصویر را بهبود بخشید
در مثال فوق احساس میشود پخش کردن مقادیر به صورت یکنواخت بهترین انتخاب برای همساز سازی نبوده
به طور مثال حدس زده می شود اگر مقادیر در بازه محدود تری پخش می شدند (مثلا صفر تا ۱۵۰) کیفیت تصویر به تصویر مطلوب نزدیک تر میشد
زیرا در تصویر همسان سازی شده مشاهده می شود که قسمت هایی از تصویر بعضا زیادی به سفید مایل شده اند
