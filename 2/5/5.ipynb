# تمرین ۵ .

آ) در سری فوریه یک تابع متناوب را به مجموع سینوس ها و کسینوس ها تجزیه می کنیم
اما در تبدیل فوریه متناوب بودن تابع شرط نیست و می توان با بی نهایت گرفتن دوره تناوب تبدیل را محاسبه کرد
ورودی ها و خروجی ها
در صورتی که خروجی مظلوب گسسته باشد می توان از هر دو ورودی پیوسته و گسسته استفاده کرد  (در صورت پیوسته بودن ورودی می توان نمونه برداری انجام داد و از ورودی پیوسته داده های گسسته استخراج کرد)
اما در صورتی که خروجی مورد نظر پیوسته باشد تنها می توان از ورودی پیوسته استفاده کرد دلیل این امر نیز تا حدی واضح است یک سیگنال پیوسته مقادیر بی نهایتی داده در خود ذخیره دارد اگر خروجی گسسته برای این سیگنال محاسبه کنیم میزانی از داده را از دست داده ایم که قابل بازیابی نیست

ب)این الگوریتم برای محاسبه سریع تر تبدیل فوریه به کار می رود
در صورت یکسان بودن ورودی تبدیل فوریه گسسته و تبدیل فوریه سریع مقادیر خروجی یکسانی تولید می کنند
این در حالی است که الگوریتم فوریه سریع از لحاظ الگوریتمی از مرتبه 
n log(n) 
محاسبات را انجام میدهد در حالی که تبدیل فوریه گسسته این کار را در 
n^2
انجام می دهد
تفاوت این دو تبدیل در محاسبات جبری و نحوه پیاده سازی الگوریتم است که باعث می شود این اختلاف در سرعت محاسبه ایجاد شود
