# تمرین ۱ .تصاویر پایه DFT
## بخش ب

در ابندا به نظر می رسد هر کدام از تصاویر پایه که با تصویر اصلی شباهت ساختاری بیشتری داشته باشد صاحب ضریب بیشتری نیز خواهد بود
این موضوع  تا حد زیادی درست است
اما نکته قابل توجه این است که وزن یک تصویر پایه حاصل ضرب نقطه ای آن در تصویر است و بدیهی است که به طور کلی هر چقد تصویر پایه به سفید نزدیک تر باشد، مستقل از تصویر اصلی، وزن آن بیشتر خواهد بود
به عنوان مثال تصویر پایه کامپوننت دی سی (مختصات ۰ و ۰) تصویری تماما سفید است و بیشترین وزن را همیشه به خود اختصاص می دهد
موضوع این است که وجود سیاهی در تصویر پایه و انطباق آن با تصویر اصلی لزوما به معنای بالا تر بودن ضریب آن نیست
هر چند مکان هایی که در تصویر اصلی سیاه هستند نشان گر آن است که این مکان ها در تصویر پایه موجب افزایش ضریب نخواهند شد

به طور کلی و خلاصه بالا بودن ضریب یک تصویر پایه فقط و فقط به انطباق قسمت های روشن آن با تصویر اصلی وابسته است
سیاه یا سفید بودن محل هایی که در تصویر اصلی سیاه هستند اهمیت چندانی ندارد
علاوه بر این موضوع علامت اجزای تصویر پایه مد نظر نیز حائز اهمیت است
تصاویر پایه ای که کل اجزای آن به طور یکپارچه هم علامت هستند قاعدتا ضریب بالاتری خواهند داشت
